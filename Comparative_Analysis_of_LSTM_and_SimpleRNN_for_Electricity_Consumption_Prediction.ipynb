```python
6# LANGKAH 1: Upload File ke Colab
from google.colab import files
import os

uploaded = files.upload()

filename = None
for fname in os.listdir():
    if fname.endswith('.csv') and 'Tetuan' in fname:
        filename = fname
        break

# LANGKAH 2: Impor Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from keras.models import Sequential
from keras.layers import LSTM, SimpleRNN, Dense
from keras.callbacks import EarlyStopping

# LANGKAH 3: Load Dataset & EDA
df = pd.read_csv(filename)
df.columns = [c.strip().replace('  ', ' ') for c in df.columns]
df['DateTime'] = pd.to_datetime(df['DateTime'])
df.set_index('DateTime', inplace=True)
df = df.dropna()

print("=== Deskripsi Dataset ===")
print(df.describe())

df.plot(subplots=True, figsize=(12, 10), title='Time Series Plots')
plt.tight_layout()
plt.show()

# LANGKAH 4: Korelasi Fitur
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# LANGKAH 5: Fitur & Target
target = 'Zone 1 Power Consumption'
features = ['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows']

print(f"\n=== Feature Selection ===\nFitur yang digunakan: {features}")
print(f"Target yang diprediksi: {target}")

data = df[features + [target]]
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)

# LANGKAH 6: Framing Time Series
def create_dataset(data, look_back=10):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back, :-1])
        y.append(data[i + look_back, -1])
    return np.array(X), np.array(y)

X, y = create_dataset(scaled)
print(f"\nDataset shape setelah framing: X={X.shape}, y={y.shape}")

# LANGKAH 7: KFold + LSTM
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mae_list, rmse_list, r2_list = [], [], []
fold = 1

print("\n=== EVALUASI LSTM ===")
for train_idx, test_idx in kf.split(X):
    print(f"\n=== Fold {fold} ===")
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    print(f"Jumlah data training: {len(X_train)}, testing: {len(X_test)}")

    model = Sequential([
        LSTM(64, input_shape=(X.shape[1], X.shape[2])),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    history = model.fit(X_train, y_train,
                        validation_split=0.2,
                        epochs=100,
                        batch_size=32,
                        callbacks=[early_stop],
                        verbose=0)

    print(f"Model LSTM selesai pada epoch ke-{len(history.history['loss'])}")

    y_pred = model.predict(X_test).flatten()

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    mae_list.append(mae)
    rmse_list.append(rmse)
    r2_list.append(r2)

    print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}, R²: {r2:.2f}")

    if fold == 5:
        plt.figure(figsize=(10, 5))
        plt.plot(y_test[:200], label='Actual')
        plt.plot(y_pred[:200], label='Predicted (LSTM)')
        plt.title('LSTM Prediction vs Actual (Fold 5)')
        plt.xlabel("Index")
        plt.ylabel("Normalized Value")
        plt.legend()
        plt.tight_layout()
        plt.show()

    fold += 1

print("\n=== RATA-RATA LSTM (5-FOLD) ===")
print(f"MAE rata-rata:  {np.mean(mae_list):.2f}")
print(f"RMSE rata-rata: {np.mean(rmse_list):.2f}")
print(f"R² rata-rata:   {np.mean(r2_list):.2f}")

# LANGKAH 8: Evaluasi SimpleRNN
print("\n=== EVALUASI SimpleRNN ===")
mae_rnn, rmse_rnn, r2_rnn = [], [], []
fold = 1

for train_idx, test_idx in kf.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    model_rnn = Sequential([
        SimpleRNN(64, input_shape=(X.shape[1], X.shape[2])),
        Dense(1)
    ])
    model_rnn.compile(optimizer='adam', loss='mse')
    early_stop_rnn = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    history_rnn = model_rnn.fit(X_train, y_train,
                                validation_split=0.2,
                                epochs=100,
                                batch_size=32,
                                callbacks=[early_stop_rnn],
                                verbose=0)

    print(f"Model SimpleRNN selesai pada epoch ke-{len(history_rnn.history['loss'])}")

    y_pred_rnn = model_rnn.predict(X_test).flatten()

    mae_rnn.append(mean_absolute_error(y_test, y_pred_rnn))
    rmse_rnn.append(np.sqrt(mean_squared_error(y_test, y_pred_rnn)))
    r2_rnn.append(r2_score(y_test, y_pred_rnn))

    print(f"MAE: {mae_rnn[-1]:.2f}, RMSE: {rmse_rnn[-1]:.2f}, R²: {r2_rnn[-1]:.2f}")

    if fold == 5:
        plt.figure(figsize=(10, 5))
        plt.plot(y_test[:200], label='Actual')
        plt.plot(y_pred_rnn[:200], label='Predicted (SimpleRNN)')
        plt.title('SimpleRNN Prediction vs Actual (Fold 5)')
        plt.xlabel("Index")
        plt.ylabel("Normalized Value")
        plt.legend()
        plt.tight_layout()
        plt.show()

    fold += 1

print("\n=== RATA-RATA SimpleRNN (5-FOLD) ===")
print(f"MAE rata-rata:  {np.mean(mae_rnn):.2f}")
print(f"RMSE rata-rata: {np.mean(rmse_rnn):.2f}")
print(f"R² rata-rata:   {np.mean(r2_rnn):.2f}")

# === EVALUASI AKHIR  ===
print("\n=== KESIMPULAN AKHIR EVALUASI MODEL ===")

if np.mean(r2_list) > np.mean(r2_rnn):
    winner = "LSTM"
else:
    winner = "SimpleRNN"

print(f"Setelah dilakukan evaluasi menggunakan 5-Fold Cross Validation, model {winner} menunjukkan performa yang lebih baik secara keseluruhan.")
print("Hal ini terlihat dari nilai rata-rata metrik evaluasi berikut:")

print(f"- MAE LSTM      : {np.mean(mae_list):.2f}")
print(f"- MAE SimpleRNN : {np.mean(mae_rnn):.2f}")
print(f"- RMSE LSTM     : {np.mean(rmse_list):.2f}")
print(f"- RMSE SimpleRNN: {np.mean(rmse_rnn):.2f}")
print(f"- R² LSTM       : {np.mean(r2_list):.2f}")
print(f"- R² SimpleRNN  : {np.mean(r2_rnn):.2f}")

print(f"\nDengan demikian, model {winner} direkomendasikan untuk digunakan dalam memprediksi konsumsi daya listrik berdasarkan fitur cuaca, karena memberikan hasil prediksi yang lebih akurat dan stabil.")
```
