```python
# LANGKAH 1: Upload File ke Colab
from google.colab import files
import os

uploaded = files.upload()
filename = None
for fname in os.listdir():
    if fname.endswith('.csv') and 'Tetuan' in fname:
        filename = fname
        break

# LANGKAH 2: Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from keras.models import Sequential
from keras.layers import LSTM, SimpleRNN, Dense, Dropout
from keras.callbacks import EarlyStopping

# LANGKAH 3: Load Dataset & Feature Engineering
df = pd.read_csv(filename)
df.columns = [c.strip().replace('  ', ' ') for c in df.columns]
df['DateTime'] = pd.to_datetime(df['DateTime'])
df.set_index('DateTime', inplace=True)
df = df.dropna()

df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek
df['month'] = df.index.month

print("=== DESKRIPSI DATA ===")
print(df.describe())

# === Heatmap Fitur Awal ===
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Matrix - Semua Fitur")
plt.tight_layout()
plt.show()

# === Feature Selection ===
target = 'Zone 1 Power Consumption'
excluded = ['Zone 2 Power Consumption', 'Zone 3 Power Consumption']
available = df.drop(columns=excluded + [target]).columns
correlation = df[available].corrwith(df[target]).abs().sort_values(ascending=False)
features = correlation.index[:8].tolist()

print("\n=== Top 8 Fitur Terpilih Berdasarkan Korelasi ===")
print(correlation[:8])

# === Preprocessing ===
X_raw = df[features]
y_raw = df[[target]]

feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

X_scaled = feature_scaler.fit_transform(X_raw)
y_scaled = target_scaler.fit_transform(y_raw)

data_scaled = np.hstack((X_scaled, y_scaled))

# === Framing Time Series ===
def create_dataset(data, look_back=30):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back, :-1])
        y.append(data[i + look_back, -1])
    return np.array(X), np.array(y)

X, y = create_dataset(data_scaled, look_back=30)
print(f"\nDataset shape: X={X.shape}, y={y.shape}")

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# === EVALUASI LSTM ===
print("\n=== EVALUASI LSTM ===")
mae_lstm, rmse_lstm, mse_lstm, r2_lstm = [], [], [], []
fold = 1

for train_idx, test_idx in kf.split(X):
    print(f"\nFold {fold} - LSTM")
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    model_lstm = Sequential([
        LSTM(64, return_sequences=True, activation='tanh', input_shape=(X.shape[1], X.shape[2])),
        Dropout(0.2),
        LSTM(32, activation='tanh'),
        Dropout(0.2),
        Dense(1)
    ])
    model_lstm.compile(optimizer='adam', loss='mse')
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    history = model_lstm.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32,
                             callbacks=[early_stop], verbose=0)

    print(f"Model LSTM berhenti di epoch: {len(history.history['loss'])}")

    y_pred = model_lstm.predict(X_test).flatten()
    y_test_inv = target_scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    y_pred_inv = target_scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()

    mse_lstm.append(mean_squared_error(y_test_inv, y_pred_inv))
    mae_lstm.append(mean_absolute_error(y_test_inv, y_pred_inv))
    rmse_lstm.append(np.sqrt(mse_lstm[-1]))
    r2_lstm.append(r2_score(y_test_inv, y_pred_inv))

    print(f"MSE: {mse_lstm[-1]:.2f}, RMSE: {rmse_lstm[-1]:.2f}, MAE: {mae_lstm[-1]:.2f}, R²: {r2_lstm[-1]:.2f}")
    fold += 1

# === EVALUASI SimpleRNN ===
print("\n=== EVALUASI SimpleRNN ===")
mae_rnn, rmse_rnn, mse_rnn, r2_rnn = [], [], [], []
fold = 1

for train_idx, test_idx in kf.split(X):
    print(f"\nFold {fold} - SimpleRNN")
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    model_rnn = Sequential([
        SimpleRNN(64, return_sequences=True, activation='tanh', input_shape=(X.shape[1], X.shape[2])),
        Dropout(0.2),
        SimpleRNN(32, activation='tanh'),
        Dropout(0.2),
        Dense(1)
    ])
    model_rnn.compile(optimizer='adam', loss='mse')
    early_stop_rnn = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    history = model_rnn.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32,
                            callbacks=[early_stop_rnn], verbose=0)

    print(f"Model RNN berhenti di epoch: {len(history.history['loss'])}")

    y_pred_rnn = model_rnn.predict(X_test).flatten()
    y_test_inv = target_scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    y_pred_rnn_inv = target_scaler.inverse_transform(y_pred_rnn.reshape(-1, 1)).flatten()

    mse_rnn.append(mean_squared_error(y_test_inv, y_pred_rnn_inv))
    mae_rnn.append(mean_absolute_error(y_test_inv, y_pred_rnn_inv))
    rmse_rnn.append(np.sqrt(mse_rnn[-1]))
    r2_rnn.append(r2_score(y_test_inv, y_pred_rnn_inv))

    print(f"MSE: {mse_rnn[-1]:.2f}, RMSE: {rmse_rnn[-1]:.2f}, MAE: {mae_rnn[-1]:.2f}, R²: {r2_rnn[-1]:.2f}")
    fold += 1

# === KESIMPULAN AKHIR ===
print("\n=== PERBANDINGAN AKHIR (RATA-RATA DARI 5 FOLD) ===")
print(">> LSTM")
print(f"MSE  : {np.mean(mse_lstm):.2f}")
print(f"RMSE : {np.mean(rmse_lstm):.2f}")
print(f"MAE  : {np.mean(mae_lstm):.2f}")
print(f"R²   : {np.mean(r2_lstm):.2f}")

print("\n>> SimpleRNN")
print(f"MSE  : {np.mean(mse_rnn):.2f}")
print(f"RMSE : {np.mean(rmse_rnn):.2f}")
print(f"MAE  : {np.mean(mae_rnn):.2f}")
print(f"R²   : {np.mean(r2_rnn):.2f}")

winner = "LSTM" if np.mean(r2_lstm) > np.mean(r2_rnn) else "SimpleRNN"
print(f"\nModel yang lebih direkomendasikan: **{winner}**")

# === FORECASTING FUNCTION ===
def forecasting_next_steps(model, data_scaled, look_back=30, n_steps=50):
    forecast = []
    input_seq = data_scaled[-look_back:, :-1]
    for _ in range(n_steps):
        input_reshaped = input_seq.reshape(1, look_back, input_seq.shape[1])
        pred_scaled = model.predict(input_reshaped, verbose=0)
        pred_inv = target_scaler.inverse_transform(pred_scaled)[0][0]
        forecast.append(pred_inv)
        new_row = np.append(input_seq[-1][:-1], pred_scaled)
        input_seq = np.vstack([input_seq[1:], new_row])
    return forecast

# === HASIL FORECASTING (TEXT ONLY) ===
forecast_lstm = forecasting_next_steps(model_lstm, data_scaled)
forecast_rnn = forecasting_next_steps(model_rnn, data_scaled)

print("\n=== Hasil Forecasting 50 Time Steps Ke Depan ===")
print("\n>> Forecasting LSTM (50 langkah):")
print(forecast_lstm)

print("\n>> Forecasting SimpleRNN (50 langkah):")
print(forecast_rnn)
```
